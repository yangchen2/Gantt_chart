# This Gantt chart was built using: https://github.com/gibsramen/ganttly

```python
from pandas import Timestamp
from ganttly import Task
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from ganttly import Gantt
import matplotlib.patches as mpatches
```

```python
staph_competition= Task(
    name="Nakatsuji et al., Cell Reports",
    start=Timestamp("2022/08/01"),
    end=Timestamp("2022/11/01"),
    PhD__Progress="Chapter 4: Atopic Dermatitis",
)

plasmid = Task(
    name="O'Neill ... Chen et al., J. Invest. Dermatol.",
    start=Timestamp("2022/09/01"),
    end=Timestamp("2022/12/30"),
    PhD__Progress="Chapter 3: Acne",
)

review = Task(
    name="*Chen et al., Frontiers Immuno",
    start=Timestamp("2022/09/01"),
    end=Timestamp("2023/04/04"),
    PhD__Progress="Chapter 1: Skin Microbiome",
)

skin_gut = Task(
    name="Dokoshi, *Chen et al., Nature Comms",
    start=Timestamp("2022/12/01"),
    end=Timestamp("2024/04/01"),
    PhD__Progress="Chapter 2: Skin-Gut Axis",
)

host_depletion = Task(
    name="Guccione et al., Nature Comms",
    start=Timestamp("2024/04/01"),
    end=Timestamp("2025/01/18"),
    PhD__Progress="Chapter 5: Additional Co-authors",
)

acne_loreal = Task(
    name="*Chen et al., TBD",
    start=Timestamp("2023/12/01"),
    end=Timestamp("2025/10/01"),
    PhD__Progress="Chapter 3: Acne",
)

ad_crosstrans = Task(
    name="*Chen et al., TBD",
    start=Timestamp("2024/07/01"),
    end=Timestamp("2026/03/01"),
    PhD__Progress="Chapter 4: Atopic Dermatitis",
)

psoriasis = Task(
    name="De Pessemier et al., JID",
    start=Timestamp("2024/02/01"),
    end=Timestamp("2024/10/27"),
    PhD__Progress="Chapter 5: Additional Co-authors",
)

trpca = Task(
    name="Myers et al., Nature Methods",
    start=Timestamp("2024/10/01"),
    end=Timestamp("2025/07/01"),
    PhD__Progress="Chapter 5: Additional Co-authors",
)

psy_stress = Task(
    name="Hung et al., Science Immunology",
    start=Timestamp("2024/08/01"),
    end=Timestamp("2025/03/30"),
    PhD__Progress="Chapter 5: Additional Co-authors",
)

diff_ab = Task(
    name="Patel, Chen et al., TBD",
    start=Timestamp("2024/11/01"),
    end=Timestamp("2025/08/30"),
    PhD__Progress="Chapter 5: Additional Co-authors",
)
```

```python
g = Gantt()
g.add_tasks([
    staph_competition,
    plasmid, 
    review,
    skin_gut,
    acne_loreal,
    psoriasis,
    host_depletion,
    ad_crosstrans,
    psy_stress,
    trpca,
    diff_ab
])

palette = dict(zip(
    ["Chapter 1: Skin Microbiome", 
     "Chapter 2: Skin-Gut Axis", 
     "Chapter 3: Acne", 
     "Chapter 4: Atopic Dermatitis"],
    sns.color_palette("Set2", 4)
))
palette["Chapter 5: Additional Co-authors"] = "lightgrey"

# Plot once
g.plot(
    frequency="month",
    interval=3,
    color_tag="PhD__Progress",
    palette=palette,
)

plt.xticks(rotation=45)
plt.title("PhD Progress by Publication Timeline", fontsize=16)
plt.axvline(x=Timestamp("2025/06/10"), color='black', linestyle='--')

# Create and add legend
legend_patches = [
    mpatches.Patch(color=color, label=chapter) 
    for chapter, color in palette.items()
]
plt.legend(
    handles=legend_patches,
    loc='center left',
    bbox_to_anchor=(1.02, 0.5),
    title="Thesis Chapters"
)

# Adjust layout to leave space for the legend
plt.tight_layout(rect=[0, 0, 0.85, 1])

# Save figure
plt.savefig('gant_chart_advancement-20250610.png', dpi=600, bbox_inches='tight')
```
