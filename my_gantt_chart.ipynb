# This Gantt chart was built using: https://github.com/gibsramen/ganttly

```python
from pandas import Timestamp
from ganttly import Task
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from ganttly import Gantt
```

```python
review_writing = Task(
    name="Manuscript Writing",
    start=Timestamp("2022/10/01"),
    end=Timestamp("2023/01/01"),
    PhD__Progress="Review",
)
review_submission = Task(
    name="Submission",
    start=Timestamp("2023/01/01"),
    end=Timestamp("2023/01/26"),
    PhD__Progress="Review",
)
review_revisions = Task(
    name="Journal Revision",
    start=Timestamp("2023/01/26"),
    end=Timestamp("2023/04/04"),
    PhD__Progress="Review",
)
skin_gut_analysis = Task(
    name="Analysis",
    start=Timestamp("2022/12/01"),
    end=Timestamp("2023/12/01"),
    PhD__Progress="Skin-Gut Axis",
)
skin_gut_submission = Task(
    name="Submission",
    start=Timestamp("2023/12/01"),
    end=Timestamp("2024/01/01"),
    PhD__Progress="Skin-Gut Axis",
)
acne_analysis = Task(
    name="Analysis",
    start=Timestamp("2023/08/01"),
    end=Timestamp("2024/07/01"),
    PhD__Progress="Acne Multi-omics",
)
acne_writing = Task(
    name="Manuscript Writing",
    start=Timestamp("2024/07/01"),
    end=Timestamp("2024/09/01"),
    PhD__Progress="Acne Multi-omics",
)
acne_submission = Task(
    name="Submission",
    start=Timestamp("2024/09/01"),
    end=Timestamp("2024/10/01"),
    PhD__Progress="Acne Multi-omics",
)
ad_analysis = Task(
    name="Analysis",
    start=Timestamp("2023/07/01"),
    end=Timestamp("2024/12/01"),
    PhD__Progress="Pediatric AD",
)
ad_writing = Task(
    name="Manuscript Writing",
    start=Timestamp("2024/12/01"),
    end=Timestamp("2025/03/01"),
    PhD__Progress="Pediatric AD",
)
ad_submission = Task(
    name="Submission",
    start=Timestamp("2025/03/01"),
    end=Timestamp("2025/04/01"),
    PhD__Progress="Pediatric AD",
)
qadabra_software = Task(
    name="Software Development",
    start=Timestamp("2023/03/01"),
    end=Timestamp("2023/08/01"),
    PhD__Progress="QADABRA",
)
qadabra_benchmarking = Task(
    name="Analysis",
    start=Timestamp("2023/08/01"),
    end=Timestamp("2024/08/01"),
    PhD__Progress="QADABRA",
)
qadabra_writing = Task(
    name="Manuscript Writing",
    start=Timestamp("2024/08/01"),
    end=Timestamp("2024/12/01"),
    PhD__Progress="QADABRA",
)
qadabra_submission = Task(
    name="Submission",
    start=Timestamp("2024/12/01"),
    end=Timestamp("2025/01/01"),
    PhD__Progress="QADABRA",
)
```

```python
g = Gantt()
g.add_tasks([review_writing, 
             review_submission, 
             review_revisions, 
             skin_gut_analysis,
             skin_gut_submission,
             acne_analysis,
             acne_writing,
             acne_submission,
             ad_analysis,
             ad_writing,
             ad_submission,
             qadabra_software,
             qadabra_benchmarking,
             qadabra_writing,
             qadabra_submission])

palette = dict(zip(["Review", "Skin-Gut Axis", "Acne Multi-omics", "Pediatric AD", "QADABRA"], sns.color_palette("plasma", 5)))
g.plot(
    frequency="month",
    interval=3,
    color_tag="PhD__Progress",
    palette=palette,
)

plt.xticks(rotation=45)
plt.xticks(rotation=45)

plt.axvline(x=Timestamp("2024/06/01"), color='black', linestyle='--')

plt.axhspan(0, 3, facecolor='purple', alpha=0.1)
plt.axhspan(3, 5, facecolor='pink', alpha=0.1)
plt.axhspan(5, 8, facecolor='red', alpha=0.1)
plt.axhspan(8, 11, facecolor='orange', alpha=0.1)
plt.axhspan(8, 15, facecolor='yellow', alpha=0.1)

plt.show()
```
